## Task 1:-

#### Problem Statements:-

Print no. of matches won by each team based on year wise (each year as one column)

```python
import pandas as pd
def data(filename):
    df=pd.read_csv(filename)
    return df 
data('DataFile/Match.csv')
df=data('DataFile/Match.csv')
```

```python
import numpy as np
def unique(l1,l2):  
    u,n=(np.unique(l1),np.unique(l2)) 
    data = {'NAME':n} 
    dff = pd.DataFrame(data)
    for i in range(0, len(u)):
        l3=[]
        for k in range (0,len(n)):
            sum=0
            for j in df.values:
                if(j[10]==n[k] and j[5]==u[i]):
                    sum=sum+1
            l3.append(sum)
        dff[u[i]] = l3
    return(dff)
l1,l2=[],[]
for i in df.values:
    l1.append(i[5])
    l2.append(i[10])
unique(l1,l2)
```

#### Problem Statements:-

Print  ManOfMach count of each player in  Hyderabad stadium

```python
import numpy as np  
def unique(l1):
    u,l2=[],[]
    x = np.array(l1) 
    u=(np.unique(x))
    for i in range(0,len(u)):
        sum=0
        for j in df.values:
            if(u[i]==j[14]):
                sum=sum+1
        l2.append(sum)
    data = {'NAME':u,
           'Count':l2} 
    dff = pd.DataFrame(data)
    return dff
l1=[]
for i in df.values:
    if(i[7]=='Hyderabad'):
        l1.append(i[14])
unique(l1)
```
